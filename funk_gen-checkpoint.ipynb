# Центральная предельная теорема своими руками

Импортируем необходимые модули

```python
import scipy.stats as sts
import numpy as np
from math import sqrt
from matplotlib import pylab as plt
%matplotlib inline
```

Выбираем показательное распределение (https://ru.wikipedia.org/wiki/Экспоненциальное_распределение) с параметором лямбда = 1
Генерируем выборку объемом 1000.
Строим гистограмму и теоритическую плотность.

```python
expon_rv = sts.expon()
sample = expon_rv.rvs(1000)
plt.hist(sample,bins=20,normed=True)
plt.xlabel('$x$')
plt.ylabel('number of samples')
x = np.linspace(0,10,100)
plt.plot(x,expon_rv.pdf(x), label = 'theoretical pdf')
plt.legend()
```

## Гистограмма распределения выборочных средних при объеме выборки  n = 5
(sample  - массив выборочных средних для выборок объема n из показательного распределения, размер массива = 1000).
Для показательного распределения $${\mathbb  {E}}[X]={\frac  {1}{\lambda }}$$
$$\operatorname {D}[X]={\frac  {1}{\lambda ^{2}}}$$
При лямбда = 1, матожидание и дисперсия = 1. Характеристики нормального распределения в ЦПТ (1, sqrt(1/5) = 0.447).

```python
sample = [expon_rv.rvs(5).mean() for x in np.arange(0,1000)]
plt.hist(sample,bins=25,normed=True,label='n = 5')
x = np.linspace(0,3,100)
plt.plot(x, sts.norm(1,sqrt(1/5.)).pdf(x), label='theoretical pdf')
plt.xlabel('$x$')
plt.ylabel('number of samples')
plt.legend()
```

## Гистограмма распределения выборочных средних при объеме выборки  n = 10
Характеристики нормального распределения в ЦПТ (1, sqrt(1/10) = 0.316).

```python
sample = [expon_rv.rvs(10).mean() for x in np.arange(0,1000)]
plt.hist(sample,bins=25,normed=True,label='n = 10')
x = np.linspace(0,3,100)
plt.plot(x, sts.norm(1,sqrt(1/10.)).pdf(x), label='theoretical pdf')
plt.xlabel('$x$')
plt.ylabel('number of samples')
plt.legend()
```

## Гистограмма распределения выборочных средних при объеме выборки n = 50
Характеристики нормального распределения в ЦПТ (1, sqrt(1/50) = 0.141).

```python
sample = [expon_rv.rvs(50).mean() for x in np.arange(0,1000)]
plt.hist(sample,bins=25,normed=True,label='n = 50')
x = np.linspace(0.5,2,100)
plt.plot(x, sts.norm(1,sqrt(1/50.)).pdf(x), label='theoretical pdf')
plt.xlabel('$x$')
plt.ylabel('number of samples')
plt.legend()
```
