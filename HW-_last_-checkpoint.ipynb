## Распределение хи-квадрат

```python
#подключаем необходимые библиотеки

%matplotlib inline
from scipy.stats import ncx2
import matplotlib.pyplot as plt
```

```python
df, nc = 21, 0.01 #(здесь и ниже) описываем параметры для построения графика
```

```python
x = np.linspace(ncx2.ppf(0.001, df, nc), ncx2.ppf(0.999, df, nc), 100)
```

```python
rv = ncx2(df, nc)
```

```python
r = ncx2.rvs(df, nc, size=1000) #создаем выборку объема 1000
```

```python
plt.ylabel('number of samples') #строим график
plt.xlabel('$x$')
plt.hist(r, normed=True, histtype='stepfilled', alpha=0.2)
plt.plot(x, rv.pdf(x), 'k-', lw=2, label='frozen pdf')
```

### Создадим 1000 выборок объёма 5, 10, 50.


```python
def fl(lst): # функция расчета среднего
    s = 0
    for i in lst:
        s += i        
    return s / len(lst)
```

```python
l1 = list() # создаем массив средних с n = 5
for i in range(1000):
    r1 = ncx2.rvs(df, nc, size=5)
    l1.append(fl(r1))
```

```python
plt.ylabel('number of samples')
plt.xlabel('$x$')
plt.hist(l1, normed=True, histtype='stepfilled', alpha=0.2) # строим гистограмму
```

```python
l2 = list() # создаем массив средних с n = 10 и строим гистограмму
for i in range(1000):
    r2 = ncx2.rvs(df, nc, size=10)
    l2.append(fl(r2))
    
plt.ylabel('number of samples')
plt.xlabel('$x$')
plt.hist(l2, normed=True, histtype='stepfilled', alpha=0.2)
```

```python
l3 = list() # создаем массив средних с n = 50 и строим гистограмму
for i in range(1000): 
    r3 = ncx2.rvs(df, nc, size=50)
    l3.append(fl(r3))
    
plt.ylabel('number of samples')
plt.xlabel('$x$')
plt.hist(l3, normed=True, histtype='stepfilled', alpha=0.2)
```
